```python
# Define files and concatenate with ignore_index=True

# List of CSV file names
csv_files = ["new_bangalore_cardata.csv", "new_chennai_cardata.csv", 
             "new_delhi_cardata.csv", "new_hyderabad_cardata.csv", 
             "new_jaipur_cardata.csv", "new_Kolkata_cardata.csv"]

# Concatenate all CSV files into a single DataFrame
all_city_cardata = pd.concat([pd.read_csv(file) for file in csv_files], ignore_index=True)

# Standardize column names by removing any trailing numbers (including floating points) and dots
all_city_cardata.columns = all_city_cardata.columns.str.replace(r'\.\d+$', '', regex=True)  # Remove '.1', '.2' etc.
all_city_cardata.columns = all_city_cardata.columns.str.replace(r'\d+$', '', regex=True)  # Remove trailing numbers

# Remove duplicate columns after standardizing names
all_city_cardata = all_city_cardata.loc[:, ~all_city_cardata.columns.duplicated()]

# Check the resulting DataFrame
print(all_city_cardata.shape)
all_city_cardata.head()
```
