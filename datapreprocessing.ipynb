Handling Missing Values and changing data types


```python
# dropping the columns that are not required
all_city_cardata.drop(columns=["it", "Gear Box", "Drive Type","Color"],inplace=True, errors='ignore')

# Confirm if the columns have been removed
all_city_cardata.columns
```

```python
all_city_cardata.isnull().sum() #checking for missing values
```

```python
# Convert 'Mileage' column to string, remove ' kmpl' and ' km/kg', and convert to numeric
all_city_cardata['Mileage'] = all_city_cardata['Mileage'].astype(str).str.replace(' kmpl', '').str.replace(' km/kg', '').astype(float)

# Fill with median
all_city_cardata['Mileage'].fillna(all_city_cardata['Mileage'].median(), inplace=True)
```

```python
# removing the rows with missing values
all_city_cardata.dropna(inplace=True)

all_city_cardata.isnull().sum() #checking for missing values
```

```python
all_city_cardata.head()   #Displaying the first 5 rows of the data
```

```python
all_city_cardata.shape #checking the shape of the data
```

```python
#changing the column names
all_city_cardata.rename(columns={"ft": "Fuel Type", "bt": "Body type", "oem": "Brand","transmission":"Transmission","ownerNo":"Owner"}, inplace=True)
all_city_cardata.head()
```

```python
all_city_cardata.dtypes #checking the data types of the columns
```

```python
#converting year column  to integer
all_city_cardata['modelYear'] = all_city_cardata['modelYear'].astype(int)

#converting price column to float
# Removing 'lakh' and converting to a numeric format
all_city_cardata['price'] = all_city_cardata['price'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)*100000

# converting kms driven by removing 'Kms', replacing commas, and converting to integer
all_city_cardata['Kms Driven'] = all_city_cardata['Kms Driven'].astype(str).str.replace('Kms', '').str.replace(',', '').str.strip().astype(int)
 
 #converting owner number to integer
all_city_cardata['Owner'] = all_city_cardata['Owner'].astype(int)

#removing the string 'CC' from engine  and converting to integer
all_city_cardata['Engine'] = all_city_cardata['Engine'].astype(str).str.replace('CC', '').str.strip().astype(int)

```

```python
#feature engineering,a new column 'Age' is created to calculate the age of the car
from datetime import datetime
now = datetime.now()
all_city_cardata[' Car Age'] = now.year - all_city_cardata['modelYear']

all_city_cardata.head()
```

```python
# removing outliers based on price
Q1 = all_city_cardata['price'].quantile(0.25)
Q3 = all_city_cardata['price'].quantile(0.75)
IQR = Q3 - Q1
all_city_cardata = all_city_cardata[(all_city_cardata['price'] >= (Q1 - 1.5 * IQR)) & (all_city_cardata['price'] <= (Q3 + 1.5 * IQR))]
all_city_cardata.shape
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select categorical columns to apply OneHotEncoding
categorical_features = ['Fuel Type', 'Body type', 'Transmission', 'Brand', 'model', 
                        'Insurance Validity', 'city']

df_encoded = pd.get_dummies(all_city_cardata, columns=categorical_features, drop_first=False, dtype=int)

# Convert any remaining numeric columns that should be integers
int_columns = ['Owner', 'modelYear', 'Kms Driven', 'Car Age']  # Specify columns that should be integer
for col in int_columns:
    if col in df_encoded.columns:
        df_encoded[col] = df_encoded[col].astype(int)
 

# Display the encoded DataFrame
df_encoded.head()
```
